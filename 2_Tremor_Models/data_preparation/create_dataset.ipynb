```python
import numpy as np
from pathlib import Path

# Path to one of your saved .npz files
npz_path = Path("../../../project_datasets/tremor/CrossArms/Healthy/1.npz")

# Load the .npz file
data = np.load(npz_path, allow_pickle=True)

# Print keys to see what’s inside
print("Keys in file:", list(data.keys()))

# Extract contents
signal = data["signal"]        # this should be a tuple (left, right)
label = data["label"].item()
wrist = data["wrist"].item()
subject_id = data["subject_id"].item()

# Inspect structure
print("\n--- File Info ---")
print(f"Subject ID: {subject_id}")
print(f"Label: {label} (0=Healthy, 1=Parkinson, 2=Other)")
print(f"Handedness (wrist): {wrist} (0=Left-handed, 1=Right-handed)")

# Since 'signal' is a tuple of two arrays, access each side
left_signal, right_signal = signal

print("\n--- Signal Shapes ---")
print(f"Left wrist signal shape:  {left_signal.shape}")
print(f"Right wrist signal shape: {right_signal.shape}")
```
