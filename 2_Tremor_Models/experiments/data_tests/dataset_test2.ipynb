```python
from dataset import TremorDataset

dataset = TremorDataset(
        data_path="../../project_datasets/tremor",  # Root folder with all movement subfolders
        random_seed=42
    )
    
print(f"\nDataset size: {len(dataset)}")

# Get a sample
signal, wrist, movement, label = dataset[0]
print(f"\nSample 0:")
print(f"  Signal shape: {signal.shape}")
print(f"  Wrist: {wrist.item()} ({'Left' if wrist.item() == 0 else 'Right'})")
print(f"  Movement: {movement.item()} ({dataset.get_movement_name(movement.item())})")
print(f"  Label: {label.item()} ({'Healthy' if label.item() == 0 else 'Parkinson' if label.item() == 1 else 'Other'})")

# Print distributions
print("\nClass distribution:")
for cls, count in dataset.get_class_distribution().items():
    print(f"  {cls}: {count}")

print("\nMovement distribution:")
for movement, count in dataset.get_movement_distribution().items():
    print(f"  {movement}: {count}")
```
