# 1. Load dataloaders

```python
from data.dataloader import create_tremor_dataloaders

train_loader, val_loader = create_tremor_dataloaders(
    "../project_datasets/tremor/",
    batch_size= 32,
    include_other= False,
    print_details=False,
)
```

# 2. Create models

```python
from Models.model_V3 import TremorNetGRU_V3

model = TremorNetGRU_V3()
```

# 3. Train models

```python
from training.trainer import train

train(
    model=model,
    train_dataloader=train_loader,
    val_dataloader=val_loader,
    
    model_name="TremorNetGRU_V3",
    run_name="TremorNetGRU_V3_run1",
    
    epochs = 10,
)
```
